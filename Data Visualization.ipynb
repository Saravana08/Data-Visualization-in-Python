# Creating Line Plot in Matplotli

```python
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)

y = np.sin(x)

#The values from two arrays are plotted using the plot() function.
plt.plot(x,y)

#You can set the plot title, and labels for x and y axes.
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
#The Plot viewer window is invoked by the show() function −

plt.show()
```

```python
iris = pd.read_csv(r'C:\Users\sarav\Downloads\iris.csv')
wine_reviews = pd.read_csv(r'C:\Users\sarav\Downloads\wine_reviews.csv')
```

## Creating Scatter Plot in Matplotlib 

```python
#Scatter the sepal_length against the sepal_width
plt.scatter(iris['sepal_length'],iris['sepal_width'])

#Set a title and labels
plt.title('Iris Dataset')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```

## Creating a Bar chart in Matplotlib

```python
#Count the occurrence of each class
data=wine_reviews['points'].value_counts()

#get x and y data
points=data.index
frequency=data.values

#create bar chart
plt.bar(points,frequency)

#set title and labels
plt.title('Wine Review Scores')
plt.xlabel('Points')
plt.xticks([60,90])
plt.yticks([0,3000,6000,9000,12000])
plt.ylabel('Frequency')
plt.show()
```

## Creating Histogram in Matplotlib

```python
#plot histogram
plt.hist(wine_reviews['points'])

#set title and labels
plt.title('Wine Review Scores')
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.show()
```

## Creating Box Plot in Matplotlib 

```python
#get columns to plot
columns=iris.columns.drop(['Id'])

#create x data
x_data=range(0, iris.shape[0])

plt.boxplot(iris['sepal_length'])
plt.show()
```

# Creating a Pie chart in Matplotlib

```python
from matplotlib import pyplot as plt
import numpy as np
plt.axis('equal')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.pie(students, labels = langs,autopct='%1.2f%%')
plt.show()
```
